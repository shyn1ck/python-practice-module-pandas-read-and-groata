### Задача 1. Анализ данных игровой индустрии (pivot_table)

3) Определите сумму, максимальную и минимальную величины по жанрам игр;

4) Выведите топ-5 используемых платформ по продажам;

5) Выведите топ-5 жанров по продажам;

6) Сохраните полученные результаты в `excel-файлы`

```python
import pandas as pd

data = pd.read_csv(
    "/Users/parvizjon/python-practicу-module-pandas-read-and-group-data/python-practice-module-pandas-read-and-groata/games.csv")
print("Первые 5 строк данных:\n", data.head())
print("Статистика данных:\n", data.describe())
print("Информация о данных:\n", data.info())

data.columns = data.columns.str.strip().str.lower()
print("Названия колонок:", data.columns)

data['total_sales'] = (
        data['na_sales'] +
        data['eu_sales'] +
        data['jp_sales'] +
        data['other_sales']
)

platform_sales = data.pivot_table(
    index='platform',
    values='total_sales',
    aggfunc=['sum', 'count'],
    margins=True
)
platform_sales.columns = ['total_sales', 'sales_count']
genre_stats = data.pivot_table(
    index='genre',
    values='total_sales',
    aggfunc=['sum', 'max', 'min'],
    margins=True
)
genre_stats.columns = ['total_sales', 'max_sales', 'min_sales']
top_5_platforms = platform_sales.nlargest(5, 'total_sales')
top_5_genres = genre_stats.nlargest(5, 'total_sales')

with pd.ExcelWriter(
        "/Users/parvizjon/python-practicу-module-pandas-read-and-group-data/python-practice-module-pandas-read-and-groata/game_sales_analysis_extended.xlsx") as writer:
    platform_sales.to_excel(writer, sheet_name='Platform Sales')
    genre_stats.to_excel(writer, sheet_name='Genre Stats')
    top_5_platforms.to_excel(writer, sheet_name='Top 5 Platforms')
    top_5_genres.to_excel(writer, sheet_name='Top 5 Genres')
```

### Задача 2. Продолжаем обрабатывать данные игровой индустрии (groupby)

Используя данные из файла `csv` в папке Work materials, нужно сделать следующее:

1) Изучите данные;

2) Определите сумму и кол-во продаж по каждой платформе. Продажи отображены в колонках с название колонки `sales`;

3) Определите сумму, медиану и среднее арифметическое по жанрам игр;

4) Выведите топ-5 наименее популярных платформ по продажам;

5) Выведите топ-5 наименее привлекательных жанров по продажам.

6) Сохраните полученные результаты в `excel-файлы`

```python
# Решение 
import pandas as pd

data = pd.read_csv(
    "/Users/parvizjon/python-practicу-module-pandas-read-and-group-data/python-practice-module-pandas-read-and-groata/games.csv")

data.columns = data.columns.str.strip().str.lower()

platform_sales = data.groupby('platform').agg(
    total_sales=('na_sales', 'sum'),
    eu_sales=('eu_sales', 'sum'),
    jp_sales=('jp_sales', 'sum'),
    other_sales=('other_sales', 'sum'),
    count_sales=('na_sales', 'count')
)

platform_sales['total_sales'] = platform_sales[['total_sales', 'eu_sales', 'jp_sales', 'other_sales']].sum(axis=1)
platform_sales = platform_sales[['total_sales', 'count_sales']]

print("Сумма и количество продаж по каждой платформе:\n", platform_sales)

genre_sales_stats = data.groupby('genre').agg(
    total_sales=('na_sales', 'sum'),
    eu_sales=('eu_sales', 'sum'),
    jp_sales=('jp_sales', 'sum'),
    other_sales=('other_sales', 'sum'),
    median_sales=('na_sales', 'median'),
    mean_sales=('na_sales', 'mean')
)

genre_sales_stats['total_sales'] = genre_sales_stats[['total_sales', 'eu_sales', 'jp_sales', 'other_sales']].sum(axis=1)
genre_sales_stats = genre_sales_stats[['total_sales', 'median_sales', 'mean_sales']]

print("Сумма, медиана и среднее по жанрам:\n", genre_sales_stats)

bottom_5_platforms = platform_sales.nsmallest(5, 'total_sales')
print("Топ-5 наименее популярных платформ:\n", bottom_5_platforms)

bottom_5_genres = genre_sales_stats.nsmallest(5, 'total_sales')
print("Топ-5 наименее привлекательных жанров:\n", bottom_5_genres)

with pd.ExcelWriter(
        "/Users/parvizjon/python-practicу-module-pandas-read-and-group-data/python-practice-module-pandas-read-and-groata/game_sales_analysis_extended.xlsx") as writer:
    platform_sales.to_excel(writer, sheet_name="Platform Sales")
    genre_sales_stats.to_excel(writer, sheet_name="Genre Sales Stats")
    bottom_5_platforms.to_excel(writer, sheet_name="Bottom 5 Platforms")
    bottom_5_genres.to_excel(writer, sheet_name="Bottom 5 Genres")
```

### Задача 3. Группировка и классификация данных

1) Прочтите Excel-файл находящийся в гугл-диске в папке `Data for work`.

2) Добавьте классификацию:

Добавьте новую колонку `amount_category`, которая будет указывать на категорию суммы транзакции. Классификация может быть следующей:

-   Low: сумма от 50 до 150
-   Medium: сумма от 151 до 300
-   High: сумма больше 300
    
Используйте `apply` и `lambda` функцию для создания этой классификации.

3) Сделайте группировку по amount_category и category, чтобы посчитать общую сумму (amount) для каждой категории и их подкатегории.

4) Создайте сводную таблицу (pivot table):

- Постройте сводную таблицу, которая покажет суммарные значения для каждой комбинации category и amount_category. 
- Убедитесь, что значения сводной таблицы сортируются по сумме транзакций по убыванию.

5) Выведите на экран результат 3 и 4 задачи 

```python
# Решение
import pandas as pd

file_path = "/Users/parvizjon/python-practicу-module-pandas-read-and-group-data/python-practice-module-pandas-read-and-groata/cw_task.xlsx"
data = pd.read_excel(file_path)

print(data.info())
print(data.head())


def classify_amount(amount):
    if 50 <= amount <= 150:
        return 'Low'
    elif 151 <= amount <= 300:
        return 'Medium'
    elif amount > 300:
        return 'High'
    else:
        return 'Unknown'


data['amount_category'] = data['amount'].apply(lambda x: classify_amount(x))

grouped_data = data.groupby(['amount_category', 'deb_type']).agg(
    total_amount=('amount', 'sum')
).reset_index()

print("Группировка по amount_category и deb_type:\n", grouped_data)

pivot_table = pd.pivot_table(
    data,
    values='amount',
    index='deb_type',
    columns='amount_category',
    aggfunc='sum'
).fillna(0).sort_values(by='High', ascending=False)

print("Сводная таблица:\n", pivot_table)
```
