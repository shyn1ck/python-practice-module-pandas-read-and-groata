### –ó–∞–¥–∞—á–∞ 1. –ê–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö –∏–≥—Ä–æ–≤–æ–π –∏–Ω–¥—É—Å—Ç—Ä–∏–∏ (pivot_table)

–ò—Å–ø–æ–ª—å–∑—É—è –¥–∞–Ω–Ω—ã–µ –∏–∑ —Ñ–∞–π–ª–∞ `csv` –≤ –ø–∞–ø–∫–µ Work materials, –Ω—É–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å —Å–ª–µ–¥—É—é—â–µ–µ:

1) –ò–∑—É—á–∏—Ç–µ –¥–∞–Ω–Ω—ã–µ;

2) –û–ø—Ä–µ–¥–µ–ª–∏—Ç–µ —Å—É–º–º—É –∏ –∫–æ–ª-–≤–æ –ø—Ä–æ–¥–∞–∂ –ø–æ –∫–∞–∂–¥–æ–π –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ. –ü—Ä–æ–¥–∞–∂–∏ –æ—Ç–æ–±—Ä–∞–∂–µ–Ω—ã –≤ –∫–æ–ª–æ–Ω–∫–∞—Ö —Å –Ω–∞–∑–≤–∞–Ω–∏–µ –∫–æ–ª–æ–Ω–∫–∏ `sales`;

3) –û–ø—Ä–µ–¥–µ–ª–∏—Ç–µ —Å—É–º–º—É, –º–∞–∫—Å–∏–º–∞–ª—å–Ω—É—é –∏ –º–∏–Ω–∏–º–∞–ª—å–Ω—É—é –≤–µ–ª–∏—á–∏–Ω—ã –ø–æ –∂–∞–Ω—Ä–∞–º –∏–≥—Ä;

4) –í—ã–≤–µ–¥–∏—Ç–µ —Ç–æ–ø-5 –∏—Å–ø–æ–ª—å–∑—É–µ–º—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º –ø–æ –ø—Ä–æ–¥–∞–∂–∞–º;

5) –í—ã–≤–µ–¥–∏—Ç–µ —Ç–æ–ø-5 –∂–∞–Ω—Ä–æ–≤ –ø–æ –ø—Ä–æ–¥–∞–∂–∞–º;

6) –°–æ—Ö—Ä–∞–Ω–∏—Ç–µ –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –≤ `excel-—Ñ–∞–π–ª—ã`

```python
import pandas as pd

try:
    data = pd.read_csv(
    "/Users/parvizjon/python-practic—É-module-pandas-read-and-group-data/python-practice-module-pandas-read-and-groata/games.csv")
except:
    data = pd.read_csv(r"C:\Users\EgamovA\Downloads\games.csv")

print("–ü–µ—Ä–≤—ã–µ 5 —Å—Ç—Ä–æ–∫ –¥–∞–Ω–Ω—ã—Ö:\n", data.head())
print("–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –¥–∞–Ω–Ω—ã—Ö:\n", data.describe())
print("–ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –¥–∞–Ω–Ω—ã—Ö:\n", data.info())

data.columns = data.columns.str.strip().str.lower()
print("–ù–∞–∑–≤–∞–Ω–∏—è –∫–æ–ª–æ–Ω–æ–∫:", data.columns)

data['total_sales'] = (
        data['na_sales'] +
        data['eu_sales'] +
        data['jp_sales'] +
        data['other_sales']
)

platform_sales = data.pivot_table(
    index='platform',
    values='total_sales',
    aggfunc=['sum', 'count'],
    margins=True
)
platform_sales.columns = ['total_sales', 'sales_count']
genre_stats = data.pivot_table(
    index='genre',
    values='total_sales',
    aggfunc=['sum', 'max', 'min'],
    margins=True
)
genre_stats.columns = ['total_sales', 'max_sales', 'min_sales']
top_5_platforms = platform_sales.nlargest(5, 'total_sales')
top_5_genres = genre_stats.nlargest(5, 'total_sales')

# with pd.ExcelWriter(
#         "/Users/parvizjon/python-practic—É-module-pandas-read-and-group-data/python-practice-module-pandas-read-and-groata/game_sales_analysis_extended.xlsx") as writer:
#     platform_sales.to_excel(writer, sheet_name='Platform Sales')
#     genre_stats.to_excel(writer, sheet_name='Genre Stats')
#     top_5_platforms.to_excel(writer, sheet_name='Top 5 Platforms')
#     top_5_genres.to_excel(writer, sheet_name='Top 5 Genres')
```

### –ó–∞–¥–∞—á–∞ 2. –ü—Ä–æ–¥–æ–ª–∂–∞–µ–º –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞—Ç—å –¥–∞–Ω–Ω—ã–µ –∏–≥—Ä–æ–≤–æ–π –∏–Ω–¥—É—Å—Ç—Ä–∏–∏ (groupby)

–ò—Å–ø–æ–ª—å–∑—É—è –¥–∞–Ω–Ω—ã–µ –∏–∑ —Ñ–∞–π–ª–∞ `csv` –≤ –ø–∞–ø–∫–µ Work materials, –Ω—É–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å —Å–ª–µ–¥—É—é—â–µ–µ:

1) –ò–∑—É—á–∏—Ç–µ –¥–∞–Ω–Ω—ã–µ;

2) –û–ø—Ä–µ–¥–µ–ª–∏—Ç–µ —Å—É–º–º—É –∏ –∫–æ–ª-–≤–æ –ø—Ä–æ–¥–∞–∂ –ø–æ –∫–∞–∂–¥–æ–π –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ. –ü—Ä–æ–¥–∞–∂–∏ –æ—Ç–æ–±—Ä–∞–∂–µ–Ω—ã –≤ –∫–æ–ª–æ–Ω–∫–∞—Ö —Å –Ω–∞–∑–≤–∞–Ω–∏–µ –∫–æ–ª–æ–Ω–∫–∏ `sales`;

3) –û–ø—Ä–µ–¥–µ–ª–∏—Ç–µ —Å—É–º–º—É, –º–µ–¥–∏–∞–Ω—É –∏ —Å—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ –ø–æ –∂–∞–Ω—Ä–∞–º –∏–≥—Ä;

4) –í—ã–≤–µ–¥–∏—Ç–µ —Ç–æ–ø-5 –Ω–∞–∏–º–µ–Ω–µ–µ –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º –ø–æ –ø—Ä–æ–¥–∞–∂–∞–º;

5) –í—ã–≤–µ–¥–∏—Ç–µ —Ç–æ–ø-5 –Ω–∞–∏–º–µ–Ω–µ–µ –ø—Ä–∏–≤–ª–µ–∫–∞—Ç–µ–ª—å–Ω—ã—Ö –∂–∞–Ω—Ä–æ–≤ –ø–æ –ø—Ä–æ–¥–∞–∂–∞–º.

6) –°–æ—Ö—Ä–∞–Ω–∏—Ç–µ –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –≤ `excel-—Ñ–∞–π–ª—ã`

```python
# –†–µ—à–µ–Ω–∏–µ 
import pandas as pd

# data = pd.read_csv(
#     "/Users/parvizjon/python-practic—É-module-pandas-read-and-group-data/python-practice-module-pandas-read-and-groata/games.csv")

data.columns = data.columns.str.strip().str.lower()

platform_sales = data.groupby('platform').agg(
    total_sales=('na_sales', 'sum'),
    eu_sales=('eu_sales', 'sum'),
    jp_sales=('jp_sales', 'sum'),
    other_sales=('other_sales', 'sum'),
    count_sales=('na_sales', 'count')
)

platform_sales['total_sales'] = platform_sales[['total_sales', 'eu_sales', 'jp_sales', 'other_sales']].sum(axis=1)
platform_sales = platform_sales[['total_sales', 'count_sales']]

print("–°—É–º–º–∞ –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–¥–∞–∂ –ø–æ –∫–∞–∂–¥–æ–π –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ:\n", platform_sales)

genre_sales_stats = data.groupby('genre').agg(
    total_sales=('na_sales', 'sum'),
    eu_sales=('eu_sales', 'sum'),
    jp_sales=('jp_sales', 'sum'),
    other_sales=('other_sales', 'sum'),
    median_sales=('na_sales', 'median'),
    mean_sales=('na_sales', 'mean')
)

genre_sales_stats['total_sales'] = genre_sales_stats[['total_sales', 'eu_sales', 'jp_sales', 'other_sales']].sum(axis=1)
genre_sales_stats = genre_sales_stats[['total_sales', 'median_sales', 'mean_sales']]

print("–°—É–º–º–∞, –º–µ–¥–∏–∞–Ω–∞ –∏ —Å—Ä–µ–¥–Ω–µ–µ –ø–æ –∂–∞–Ω—Ä–∞–º:\n", genre_sales_stats)

bottom_5_platforms = platform_sales.nsmallest(5, 'total_sales')
print("–¢–æ–ø-5 –Ω–∞–∏–º–µ–Ω–µ–µ –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º:\n", bottom_5_platforms)

bottom_5_genres = genre_sales_stats.nsmallest(5, 'total_sales')
print("–¢–æ–ø-5 –Ω–∞–∏–º–µ–Ω–µ–µ –ø—Ä–∏–≤–ª–µ–∫–∞—Ç–µ–ª—å–Ω—ã—Ö –∂–∞–Ω—Ä–æ–≤:\n", bottom_5_genres)

# with pd.ExcelWriter(
#         "/Users/parvizjon/python-practic—É-module-pandas-read-and-group-data/python-practice-module-pandas-read-and-groata/game_sales_analysis_extended.xlsx") as writer:
#     platform_sales.to_excel(writer, sheet_name="Platform Sales")
#     genre_sales_stats.to_excel(writer, sheet_name="Genre Sales Stats")
#     bottom_5_platforms.to_excel(writer, sheet_name="Bottom 5 Platforms")
#     bottom_5_genres.to_excel(writer, sheet_name="Bottom 5 Genres")
```

### –ó–∞–¥–∞—á–∞ 3. –ì—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∞ –∏ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏—è –¥–∞–Ω–Ω—ã—Ö

1) –ü—Ä–æ—á—Ç–∏—Ç–µ Excel-—Ñ–∞–π–ª –Ω–∞—Ö–æ–¥—è—â–∏–π—Å—è –≤ –≥—É–≥–ª-–¥–∏—Å–∫–µ –≤ –ø–∞–ø–∫–µ `Data for work`.

2) –î–æ–±–∞–≤—å—Ç–µ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏—é:

–î–æ–±–∞–≤—å—Ç–µ –Ω–æ–≤—É—é –∫–æ–ª–æ–Ω–∫—É `amount_category`, –∫–æ—Ç–æ—Ä–∞—è –±—É–¥–µ—Ç —É–∫–∞–∑—ã–≤–∞—Ç—å –Ω–∞ –∫–∞—Ç–µ–≥–æ—Ä–∏—é —Å—É–º–º—ã —Ç—Ä–∞–Ω–∑–∞–∫—Ü–∏–∏. –ö–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏—è –º–æ–∂–µ—Ç –±—ã—Ç—å —Å–ª–µ–¥—É—é—â–µ–π:

-   Low: —Å—É–º–º–∞ –æ—Ç 50 –¥–æ 150
-   Medium: —Å—É–º–º–∞ –æ—Ç 151 –¥–æ 300
-   High: —Å—É–º–º–∞ –±–æ–ª—å—à–µ 300
    
–ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ `apply` –∏ `lambda` —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è —Å–æ–∑–¥–∞–Ω–∏—è —ç—Ç–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏.

3) –°–¥–µ–ª–∞–π—Ç–µ –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫—É –ø–æ amount_category –∏ category, —á—Ç–æ–±—ã –ø–æ—Å—á–∏—Ç–∞—Ç—å –æ–±—â—É—é —Å—É–º–º—É (amount) –¥–ª—è –∫–∞–∂–¥–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –∏ –∏—Ö –ø–æ–¥–∫–∞—Ç–µ–≥–æ—Ä–∏–∏.

4) –°–æ–∑–¥–∞–π—Ç–µ —Å–≤–æ–¥–Ω—É—é —Ç–∞–±–ª–∏—Ü—É (pivot table):

- –ü–æ—Å—Ç—Ä–æ–π—Ç–µ —Å–≤–æ–¥–Ω—É—é —Ç–∞–±–ª–∏—Ü—É, –∫–æ—Ç–æ—Ä–∞—è –ø–æ–∫–∞–∂–µ—Ç —Å—É–º–º–∞—Ä–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –¥–ª—è –∫–∞–∂–¥–æ–π –∫–æ–º–±–∏–Ω–∞—Ü–∏–∏ category –∏ amount_category. 
- –£–±–µ–¥–∏—Ç–µ—Å—å, —á—Ç–æ –∑–Ω–∞—á–µ–Ω–∏—è —Å–≤–æ–¥–Ω–æ–π —Ç–∞–±–ª–∏—Ü—ã —Å–æ—Ä—Ç–∏—Ä—É—é—Ç—Å—è –ø–æ —Å—É–º–º–µ —Ç—Ä–∞–Ω–∑–∞–∫—Ü–∏–π –ø–æ —É–±—ã–≤–∞–Ω–∏—é.

5) –í—ã–≤–µ–¥–∏—Ç–µ –Ω–∞ —ç–∫—Ä–∞–Ω —Ä–µ–∑—É–ª—å—Ç–∞—Ç 3 –∏ 4 –∑–∞–¥–∞—á–∏ 

```python
# –†–µ—à–µ–Ω–∏–µ
import pandas as pd

file_path = "/Users/parvizjon/python-practic—É-module-pandas-read-and-group-data/python-practice-module-pandas-read-and-groata/cw_task.xlsx"
data = pd.read_excel(file_path)

print(data.info())
print(data.head())


def classify_amount(amount):
    if 50 <= amount <= 150:
        return 'Low'
    elif 151 <= amount <= 300:
        return 'Medium'
    elif amount > 300:
        return 'High'
    else:
        return 'Unknown'


data['amount_category'] = data['amount'].apply(lambda x: classify_amount(x))

grouped_data = data.groupby(['amount_category', 'deb_type']).agg(
    total_amount=('amount', 'sum')
).reset_index()

print("–ì—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∞ –ø–æ amount_category –∏ deb_type:\n", grouped_data)

pivot_table = pd.pivot_table(
    data,
    values='amount',
    index='deb_type',
    columns='amount_category',
    aggfunc='sum'
).fillna(0).sort_values(by='High', ascending=False)

print("–°–≤–æ–¥–Ω–∞—è —Ç–∞–±–ª–∏—Ü–∞:\n", pivot_table)
```
